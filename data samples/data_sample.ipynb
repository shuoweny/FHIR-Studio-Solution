```python
import pandas as pd
import matplotlib.pyplot as plt

# Creating a dictionary
data = {
    "Date": ['Feb 13', 'Feb 14', 'Feb 15', 'Feb 16', 'Feb 17', 'Feb 18', 'Feb 19', 'Feb 20', 'Feb 21', 'Feb 22'],
    "OxygenSaturation": [80, 84, 92, 99, 96, 98, 96, 97, 89, 98]
}

# Creating DataFrame from dictionary
df = pd.DataFrame(data)

# Converting 'Date' to datetime format
df['Date'] = pd.to_datetime(df['Date'], format='%b %d')

# Setting 'Date' as index
df.set_index('Date', inplace=True)

# Exporting to csv
df.to_csv('oxygen_saturation.csv')

# Plotting
plt.figure(figsize=(15,5))  # Increase width from 10 to 15
plt.plot(df['OxygenSaturation'])
plt.title('Oxygen Saturation Time Series')
plt.xlabel('Date')
plt.ylabel('Oxygen Saturation')
plt.grid(True)
plt.show()
```
